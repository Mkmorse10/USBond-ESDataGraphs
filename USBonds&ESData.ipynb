```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import yfinance as yf
%matplotlib inline
```

```python
US = pd.read_csv("/Users/mikayla/Desktop/USRaw.txt", index_col = 0, parse_dates=True)
ES = pd.read_csv("/Users/mikayla/Desktop/ESRaw.txt", index_col = 0, parse_dates=True)
ES.dropna(inplace=True)
US.dropna(inplace=True)
US.head()
```

```python
US['Daily Return'] = US['Close'].pct_change()
US.dropna(inplace=True)
```

```python
ES['Daily Return'] = ES['Close'].pct_change()
ES.dropna(inplace=True)
```

```python
# Lets plot the daily return percentage
US['Daily Return'].plot(figsize=(20,10), legend=True, linestyle='--', marker='o')
```

```python
#Make new columns
df = pd.DataFrame()
df['Bond Change'] = US['Daily Return']
df ['ES Change'] = ES['Daily Return']
```
